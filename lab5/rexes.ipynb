```python
import re

def ishem(filename):
    pattern = r'a[b]*'

    with open(filename, 'r', encoding="utf-8") as file:  
        for line in file:
            line = line.strip() 
            if re.fullmatch(pattern, line):
                print(f"{line} true")

filename = "row.txt"
ishem(filename)
```

```python
def iskat(filename):
    pattern = r'a[b]{4,}'  
    
    with open(filename, 'r',encoding="utf-8") as file:
        for line in file:
            line = line.strip()  
            if re.fullmatch(pattern, line):  
                print(f"{line} nashel")
          

filename = "row.txt"
iskat(filename)
```

```python
import re

def n(filename):
    with open(filename, 'r', encoding="utf-8") as file:
        for line in file:
            line = line.strip()
            new = re.sub(r" ", ":", line)
            new = re.sub(r".", ":", line)
            new = re.sub(r",", ":", line)
            print(f"result {new}")

filename = "row.txt"
n(filename)
```
